```python
h = int(input("Nhập độ cao h: "))

for i in range(1, h + 1):
    for j in range(1, 2 * h):
        # In dấu * ở cạnh trái, cạnh phải hoặc hàng đáy
        if j == h - i + 1 or j == h + i - 1 or i == h:
            print("*", end="")
        else:
            print(" ", end="")
    print()
#B4
```

```python
n = int(input())
oc = 0
ec = 0

if n == 0: # Trường hợp đặc biệt số 0
    ec = 1
else:
    temp = abs(n) # Đảm bảo hoạt động với cả số âm
    while temp > 0:
        digit = temp % 10  # Lấy chữ số cuối cùng
        if digit % 2 == 0:
            ec += 1
        else:
            oc += 1
        temp //= 10        # Bỏ chữ số cuối cùng đã kiểm tra

print("Even:", ec, "Odd:", oc)

```
